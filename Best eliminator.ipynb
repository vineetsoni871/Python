```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Pandas\\Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Gender'] = lb.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['Purchased'])
y = df['Purchased']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2,
                                                       random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
# Hyperparameter tunning
```

```python
param_grid = [
    {'penalty':['l1' , 'l2' , 'elasticnet' , 'none'],
    'C':np.logspace(-4,4,20),
    'solver':['lbfgs' , 'newton-cg' , 'libliner' , 'sag' , 'saga'],
    'max_iter':[100,1000,2500,5000]}
]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(lr , param_grid , n_jobs=-1,cv=3,verbose=True)
```

```python
best_clf = clf.fit(x_train,y_train)
```

```python
best_clf.best_estimator_
```

```python
print(f"Accuracy - : {accuracy_score(y_test,best_clf.predict(x_test))}")
```
