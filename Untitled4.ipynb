```python
# 1) label encoding
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Pandas\\insurance.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['sex'] = lb.fit_transform(df['sex'])
df['smoker'] = lb.fit_transform(df['smoker'])
df['region'] = lb.fit_transform(df['region'])
```

```python
df.head()
```

```python
x = df.drop(columns = ['charges'])
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_traibn , y_test = train_test_split(x,y,
                                                       test_size = 0.2,
                                                       random_state=42)
```

```python
x_train.head()
```

```python
np.round(x_train.describe(), 2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn , columns = x_train.columns)
```

```python
np.round(x_train_new.describe() , 2)
```

```python
# 2) Ordinal Encoding
```

```python
df = pd.read_csv("C:\\Pandas\\insurance.csv")
```

```python
df.head(3)
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df = df.drop(columns = ['age' , 'children' ])
df.head(3)
```

```python
df = df.drop(columns = ['charges' ])
df.head()
```

```python
df = df.drop(columns = ['bmi' ])
df.head()
```

```python
df['region'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['male' , 'female'],
                                 ['yes' , 'no'],
                                 ['southeast' , 'southwest' , 'northeast' , 'northwest']])
```

```python
df_sc = oe.fit_transform(df)
```

```python
df_new = pd.DataFrame(df_sc , columns = df.columns)
df_new
```
